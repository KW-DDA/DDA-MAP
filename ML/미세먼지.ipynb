```python
year = input("원하는 년도 입력: ")
path_dir = "data/미세먼지데이터/미세먼지_" + year +"/"
file_list = os.listdir(path_dir)
length = len(file_list)

for i in range(0, length):
    globals()["pm10_{}".format(i+1)] = pd.read_csv(path_dir+year+"_"+str(i+1)+".csv", encoding='cp949') #csv로 다시 읽기
```

```python
for i in range(1, length+1):
    globals()["pm10_{}".format(i)]=globals()["pm10_{}".format(i)].transpose()
    globals()["pm10_{}".format(i)]=globals()["pm10_{}".format(i)].rename(columns=globals()["pm10_{}".format(i)].iloc[0])
    globals()["pm10_{}".format(i)]=globals()["pm10_{}".format(i)] = globals()["pm10_{}".format(i)]=globals()["pm10_{}".format(i)].drop(globals()["pm10_{}".format(i)].index[0])
    globals()["pm10_{}".format(i)].reset_index(drop=True, inplace=True)
```

```python
for i in range(1, length+1):
    globals()["pm10_{}".format(i)]['년'] = year
    globals()["pm10_{}".format(i)]['월'] = i
    globals()["pm10_{}".format(i)]['일'] = range(1, len(globals()["pm10_{}".format(i)])+1)
```

```python

pm10 = pd.DataFrame(columns=['년','월','일','미세먼지','자치구'])
```

```python
for i in range(1, length+1):
    average=globals()["pm10_{}".format(i)]['서울시 평균'].to_list()
    for j in range(0, len(gu_ko)):
        temp = globals()["pm10_{}".format(i)][['년','월','일',gu_ko[j]]]
        temp = temp.rename(columns={gu_ko[j]: "미세먼지"})
        temp['미세먼지'].fillna(value=pd.Series(average), inplace=True) #결측치 서울시 평균으로 채우기
        temp['자치구'] = gu_ko[j]
        
        pm10 = pm10.append(temp)
        del temp
    
pm10.reset_index(drop=True, inplace=True)
pm10
```

```python

degree = pd.DataFrame(columns=['등급'])
```

```python
pm10 = pm10.append(degree) 
o_len = len(pm10)

for i in range(0, o_len):
    if pm10['미세먼지'][i] >= 0 and pm10['미세먼지'][i] <= 30:
        pm10['등급'][i] = '좋음'
    elif pm10['미세먼지'][i] >= 30 and pm10['미세먼지'][i] <= 80:
        pm10['등급'][i] = '보통'
    elif pm10['미세먼지'][i] >= 80 and pm10['미세먼지'][i] <= 150:
        pm10['등급'][i] = '나쁨'
    elif pm10['미세먼지'][i] >= 150:
        pm10['등급'][i] = '매우나쁨'
```

```python
pm10.drop(['미세먼지'], axis=1, inplace=True)
pm10.rename(columns={'등급': '미세먼지'}, inplace=True)
```

```python
pm10.to_csv("data/미세먼지데이터/미세먼지/"+year+"_미세먼지"+".csv", encoding='cp949', index=False) # csv로 저장
```
